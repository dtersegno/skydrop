Meteostat

Downloaded manual data from Barkerville, mear Pittsfield MA.

```python
import pandas as pd
import os
import matplotlib.pyplot as plt

data_path = '../weather-data/'
meteo_data = 'meteostat'
barker_data = 'barkerville.csv'


barker_path = os.path.join(data_path, meteo_data, barker_data)
wdf = pd.read_csv(barker_path)

wdf['time'] = pd.to_datetime(wdf['time'])


wdf.info()
```

```python
wdf.plot(x='time', y='temp')
plt.title('Barkerville temperature')
```

Use the api

```python
import datetime as dt
import meteostat
```

```python
october_mountain_latlon = [42.4085701,-73.2367577, 10.]

# Set time period
start = dt.datetime(2024, 9, 17)
end = dt.datetime(2024, 9, 26)

# Create Point for Vancouver, BC
october_mountain = meteostat.Point(*october_mountain_latlon)

# Get daily data for 2018
data = meteostat.Hourly(october_mountain, start, end)
data = data.fetch()

# # Plot line chart including average, minimum and maximum temperature
# data.plot(y=['tavg', 'tmin', 'tmax'])
# plt.show()
```

```python
data.columns
```

```python
distance = 75*1.6*1000 #meters  (75 mi times 1.6 km per mi times 1km/1km)
stations = meteostat.Stations().nearby(42.4085701, -73.2367577, distance)

stations = stations.fetch()
stations.plot(x='longitude', y = 'latitude', kind = 'scatter')
```

NOAA Weather.gov api
---

```python
noaa_api = 'https://api.weather.gov/'

import requests

sample_latlon = [42.0, 72.0]
```

```python
response = requests.get(f"{noaa_api}{sample_latlon[0]},{sample_latlon[1]}")
response.status_code
```
