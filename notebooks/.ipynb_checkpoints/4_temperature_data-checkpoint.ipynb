# Implement Meteo to get nearby temperatures

As of Sept 27, the 75-mile radius circle appears to be about centered at Worthington.

I'll just pick the Devrie Fine Art campus at [42.3825684,-72.9740431,1357m] according to Google

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import datetime as dt
import meteostat

#paths
data_path = '../weather-data/'
meteo_data = 'meteostat'

#locations
october_mountain = {"lat":42.4085701, "lon":-73.2367577}
devry = {"lat":42.3825684, "lon":-72.9740431, 'alt':1357}

#distances
diameter_sep27 = 75*1.6*1000 #meters
# radius = diameter_sep27/2
#just make the radius huge. just more data is all.
radius = diameter_sep27

# Set time period
start = dt.datetime(2024, 9, 17)
end = dt.datetime(2024, 9, 27)

# Create Point for devry
pDevry = meteostat.Point(**devry)

# Get daily data for 2018
# data = meteostat.Hourly(october_mountain, start, end)
# data = data.fetch()

# # Plot line chart including average, minimum and maximum temperature
# data.plot(y=['tavg', 'tmin', 'tmax'])
# plt.show()distance = 75*1.6*1000 #meters  (75 mi times 1.6 km per mi times 1km/1km)
stations = meteostat.Stations().nearby(42.4085701, -73.2367577, radius)

stations = stations.fetch()
stations.plot(x='longitude', y = 'latitude', kind = 'scatter')

data = meteostat.Hourly(stations, start, end)
data = data.fetch()
```

```python
data.reset_index()
```

```python
data_noindex = data.reset_index()


fig, ax = plt.subplots(figsize=(12,8))

for station in data_noindex['station'].unique():
    this_station_data = data_noindex.loc[data_noindex['station'] == station]
    # data_noindex.loc[data_noindex['station'] == station].plot(x='time', y='temp')
    ax.plot(this_station_data['time'], this_station_data['temp'], label=station)
    
plt.legend()
plt.show()
```

```python
#save this data!!
save_path = '../weather-data/meteostat/devries-centered-75mi.csv'
data_noindex.to_csv(save_path, index = None)
```
